```python
import pandas as pd
import pymongo
%matplotlib inline
```

```python
connection_local = "mongodb://localhost:27017"
local = pymongo.MongoClient(connection_local)

itens = local.noticias["noticias"].find({})
```

```python
itens_lista = list(itens)
```

```python
df = pd.DataFrame(itens_lista)
```

```python
df
# DataFrame com todos os dados do banco de dados.
```

```python
df['autor'].value_counts()

# Lista de autores e quantidade de artigos
```

```python
df['autor'].value_counts().describe()

# É possível ver os dados sobre os autores
```

```python
df['palavras'].value_counts().describe()

# É possivel ver a quantidade total de palavras e valores relacionados a quantidade de palavras em cada matéria
```

```python
df.info()
# É possível ver informações sobre cada coluna no df.
```

```python
# Existem 308 matérias no banco de dados e é possível ver que alguns campos estão faltando em algumas matérias
```

```python
# Plotando graficos
```

```python
df['autor'].value_counts().plot.barh(figsize=(12, 8), title=("Quantidade de artigos por autores"))
```

```python
df['paragrafos'].value_counts().plot.bar(figsize=(12, 8), title=("Quantidade de paragrafos por texto"))
```

```python
df['spider_name'].value_counts().plot.barh(figsize=(8, 5), title=("Quantidade de materias por spider"))
```
